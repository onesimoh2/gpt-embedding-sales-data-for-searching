```python
salesdata.csv
import csv
import json
import pandas as pd

data = {}

with open('data/') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        date = row[1]
        product = row[2]
        quantity = int(row[3])
        price = float(row[4].replace(',', ''))
        
        if date not in data:
            data[date] = []
        
        data[date].append({
            'product': product,
            'quantity': quantity,
            'price': price
        })
dataStr = json.dumps(data, indent=4)
data = json.loads(dataStr)
#print(json.dumps(data, indent=4))

objects = []
dates = []

for date in data:
    obj = data[date]
    new_dict = {date: obj}
    obj_str = json.dumps(new_dict)
    dates.append(date)
    objects.append(obj_str)

df = pd.DataFrame({'dates':dates,  'products': objects})
df['embeddings'] = 0.0
df.to_csv('data/sales_data_emb.csv', index=False)

df
```
