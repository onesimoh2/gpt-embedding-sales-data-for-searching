```python
import pandas as pd
import numpy as np
import openai

def get_embedding(text, model="text-embedding-ada-002"):
    text = text.replace("\n", " ")
    res = openai.Embedding.create(input = [text], model=model)
    emb = res['data'][0]['embedding']
    return emb

df = pd.read_csv("data/sales_data_emb.csv")
df = df.drop('embeddings', axis=1)

embedding_model = "text-embedding-ada-002"
openai.api_key = ""

df['embeddings'] = df.products.apply(lambda x: get_embedding(x, embedding_model))

df.to_csv('data/sales_data_emb.csv', index=False)
df
```
